```python
import numpy as np

weights = np.array([.7, .2, -.5])# initial weights selected random
alpha = .1 # limiter

#middle row represents walking(1) or staying(0)
streetlights = np.array([[0, 0, 1],
                         [0, 1, 1],
                         [0, 0, 1],
                         [1, 1, 1],
                         [0, 1, 1],
                         [1, 0, 1]])

walk_vs_stop = np.array([0, 1, 0, 1, 1, 0])

print(walk_vs_stop[4])
#print(input.dot(weights))

for iteration in range (40):
    error_for_all_lights = 0
    for row_index in range (len(walk_vs_stop)): # length = 6
        input = streetlights[row_index] # All 3
        goal_prediction = walk_vs_stop[row_index] #actual value provided 0 or 1 from [0, 1, 0, 1, 1, 0]
        prediction = input.dot(weights) # input * weights => matrix * vector
        error = (prediction -goal_prediction) ** 2 #square error
        error_for_all_lights += error # accumulate error
        #print("in", error_for_all_lights)
        delta = prediction - goal_prediction #difference between actual to prediction
        weights = weights -(alpha * input * delta) #weights update
        
        print("prediction: ", prediction)

    print("weights: ", weights)
    print("Error: ", error)
```
